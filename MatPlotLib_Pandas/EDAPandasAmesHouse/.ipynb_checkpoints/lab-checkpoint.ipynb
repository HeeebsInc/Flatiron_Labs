```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
```

```python
df = pd.read_csv('ames_train.csv')
df.head(3)
```

```python
# Investigate the Data
df.info()
```

```python
# Investigating Distributions using scatter_matrix
#sns.pairplot(df.dropna(), hue = 'species')
df_plot = df.select_dtypes(include = ['int32', 'int64', 'float32', 'float64'])
df_plot.boxplot(figsize = (10,10))
plt.title('Distribution of Ames Housing Data')
plt.xticks(rotation = 90)
plt.axis('tight')

#pd.plotting.scatter_matrix(df, alpha = .2)
#pd.boxplot
#plt.show()
```

```python
# Create a plot that shows the SalesPrice Distribution
plt.subplot(2,2,1)
sale_price = df['SalePrice'].value_counts()
plt.scatter(x = sale_price.index, y = sale_price)
plt.xlabel('Sale Price'); plt.ylabel('Counts'); plt.title('Distribution of Sale Price ')
plt.subplot(2,2,2)
df.boxplot(column = ['SalePrice'])
```

```python
# Create a plot that shows the LotArea Distribution
lotarea = df['LotArea'].value_counts()
plt.scatter(x = lotarea.index, y = lotarea)
plt.ylabel('Counts'); plt.xlabel('LotArea'); plt.title('LotArea Distribution')
```

```python
# Create a plot that shows the Distribution of the overall house condition
house_condition = df['OverallCond'].value_counts()
plt.scatter(x = house_condition.index, y = house_condition)
plt.xlabel('Overall Condition'); plt.ylabel('Count'); plt.title('Distribution of Overall House Condition')
```

```python
# Create a Box Plot for SalePrice

boxplot = df.boxplot(column = ['SalePrice'])
plt.title('BoxPlot SalePrice'); 
```

```python
# Perform an Exploration of home values by age

plt.scatter(x = df['YrSold'], y = df['SalePrice'])
plt.title('Home Values vs. Age'); plt.xlabel('YearSold'); plt.ylabel('Price Sold')
```
